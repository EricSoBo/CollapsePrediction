```python
# Test with real data

# Reservoir computing

import numpy as np
from pyESN import ESN
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Sunspot data

import csv

date = []
sunspots = []
with open('Sunspots.csv') as csvfile:
    reader = csv.reader(csvfile)
    reader.__next__()
    for row in reader:
        date.append(row[1])
        sunspots.append(float(row[2]))
```

```python
sunspots = np.array(sunspots)
trainlen = 2000
future = len(sunspots) - trainlen
window = 1
```

```python
def run_esn(window):
    esn = ESN(n_inputs = 1, n_outputs=1, n_reservoir=512, sparsity=0.05)
    current_set = []
    for i in range(0, future, window):
        pred_training = esn.fit(np.ones(trainlen), sunspots[i:trainlen+i])
        prediction = esn.predict(np.ones(window))
        for j in range(window):
            current_set.append(prediction[j])
    current_set = np.reshape(np.array(current_set), (-1, future))
    mse = mean_squared_error(sunspots[trainlen:trainlen+future][np.newaxis], current_set)
    print("Mean Squared Error: ", mse)
    
    return (mse, current_set)
```

```python
# Plotting the sunspot data

plt.plot(range(0, len(sunspots)), sunspots, label='Data', color='blue')
plt.title('Sunspot Data')
plt.xlabel('Months')
plt.ylabel('Sunspots')
plt.show()
```

```python
error, valid_pred = run_esn(window)

# Plotting the sunspot data and the prediction

plt.plot(range(0, len(sunspots)), sunspots, label='Data', color='blue')
plt.plot(range(trainlen, trainlen+future), valid_pred.T, label='Prediction', color='red')
lo, hi = plt.ylim()
plt.plot([trainlen, trainlen], [lo+np.spacing(1), hi-np.spacing(1)], 'k:', linewidth=4)
plt.title('Sunspot Data')
plt.xlabel('Months')
plt.ylabel('Sunspots')
plt.show()
```

```python
plt.plot(range(trainlen, len(sunspots)), sunspots[trainlen:], label='Data', color='blue')
plt.plot(range(trainlen, trainlen+future), valid_pred.T, label='Prediction', color='red')
lo, hi = plt.ylim()
plt.plot([trainlen, trainlen], [lo+np.spacing(1), hi-np.spacing(1)], 'k:', linewidth=4)
plt.title('Sunspot Data')
plt.xlabel('Months')
plt.ylabel('Sunspots')
plt.ylim(0, max(sunspots[trainlen:trainlen+future]))
plt.show()
```
